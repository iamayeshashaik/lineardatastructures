```python
# 1.   Delete the elements in an linked list whose sum is equal to zero

class ListNode:
    def __init__(self, data):
        self.val = data
        self.next = None

def make_list(v):
    head = ListNode(v[0])
    for i in range(1, len(v)):
        ptr = head
        while ptr.next != None:
            ptr = ptr.next
        ptr.next = ListNode(v[i])
    return head

def print_list(head):
    ptr = head
    print("[", end="")
    while ptr:
        print(ptr.val, end="")
        if ptr.next:
            print(", ", end="")
        ptr = ptr.next
    print("]")

class Solution:
    def removeZeroSumSublists(self, head: ListNode) -> ListNode:
        dummy = ListNode(0)
        dummy.next = head
        m = {0: dummy}
        sum = 0
        while head:
            sum += head.val
            m[sum] = head
            head = head.next
        head = dummy
        sum = 0
        while head:
            sum += head.val
            temp = m[sum]
            if temp != head:
                head.next = temp.next
            head = head.next
        return dummy.next

v1 = [1,2,3,-3,1,2]
head = make_list(v1)
ob = Solution()
print_list(ob.removeZeroSumSublists(head))
```

```python
# 2.  Reverse a linked list in groups of given size.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def rev_LL(head, group_size):
    current = head
    prev = None
    next_node = None
    count = 0
    
    # Reverse the nodes in the current group
    while current is not None and count < group_size:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1
    
    # Recursively call the function to reverse the next group
    if next_node is not None:
        head.next = rev_LL(next_node, group_size)
    
    # Return the new head of the reversed linked list
    return prev

head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)
head.next.next.next.next.next.next.next.next = Node(9)

# Reverse the linked list in groups of 2
new_head = rev_LL(head, 3)

# Print the reversed linked list
current = new_head
while current is not None:
    print(current.data, end=' ')
    current = current.next
```

```python
# 3.  Merge a linked list into another linked list at alternate positions.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_alt_pos(first, second):
    if first is None:
        return second
    
    current1 = first
    current2 = second
    
    # Traverse both linked lists simultaneously and merge nodes
    while current1 is not None and current2 is not None:
        next1 = current1.next
        next2 = current2.next
        
        current1.next = current2
        current2.next = next1
        
        current1 = next1
        current2 = next2
    
    return first

# Create the first linked list
first = Node(1)
first.next = Node(3)
first.next.next = Node(5)

# Create the second linked list
second = Node(2)
second.next = Node(4)
second.next.next = Node(6)

# Merge the second linked list into the first linked list at alternate positions
merged_head = merge_alt_pos(first, second)

# Print the merged linked list
current = merged_head
while current is not None:
    print(current.data, end=' ')
    current = current.next
```

```python
# 4.  In an array, Count Pairs with given sum

def count_pairs(arr, target_sum):
    count = 0
    seen = set()
    
    for num in arr:
        complement = target_sum - num
        if complement in seen:
            count += 1
        seen.add(num)
    
    return count


arr = [1, 2, 3, 4, 5]
target_sum = 7
count = count_pairs(arr, target_sum)
print("Count Pairs with given sum", target_sum, "in", arr, ":", count)
```

```python
# 5.  Find duplicates in an array

def find_dup(arr):
    seen = set()
    duplicates = set()
    
    for num in arr:
        if num in seen:
            duplicates.add(num)
        else:
            seen.add(num)
    
    return duplicates

arr = [2,3,4,5,3,4,5,6]
duplicates = find_dup(arr)
print("Duplicates in arr is :", duplicates)
```

```python
# 6.  Find the Kth largest and Kth smallest number in an array

def larg_smal(arr, k):
    arr.sort()
    kth_smallest = arr[k-1]
    kth_largest = arr[-k]
    return kth_smallest, kth_largest

arr = [1,2,4,5,6,7,8]
k = 2

kth_smallest, kth_largest = larg_smal(arr, k)

print("Kth smallest and kth largest numbers in", arr, "for k =", k, ":", kth_smallest, kth_largest)
```

```python
# 7.  Move all the negative elements to one side of the array

def negtv_move(arr):
    n = len(arr)
    j = 0
    
    for i in range(n):
        if arr[i] < 0:
            arr[i], arr[j] = arr[j], arr[i]
            j += 1
    
    return arr

arr = [1, -3, -2, 5, 4, -7,-3,5,-9,-7,-22,7,8,9,5,6,7]
arr = negtv_move(arr)

print("Array with negative elements moved to one side:", arr)
```

```python
# 8.  Reverse a string using a stack data structure

def reverse_string(string):
    stack = []
    reversed_string = ""
    
    for char in string:
        stack.append(char)
    
    while len(stack) != 0:
        reversed_string += stack.pop()
    
    return reversed_string


string = "Edyoda certi"
reversed_string = reverse_string(string)
print("Original string:", string)
print("Reversed string:", reversed_string)
```

```python
# 9.  Evaluate a postfix expression using stack

def evaluate_postfix(postfix_expression):
    stack = []
    
    for char in postfix_expression:
        if char.isdigit():
            stack.append(int(char))
        else:
            operand_2 = stack.pop()
            operand_1 = stack.pop()
            if char == '+':
                stack.append(operand_1 + operand_2)
            elif char == '-':
                stack.append(operand_1 - operand_2)
            elif char == '*':
                stack.append(operand_1 * operand_2)
            elif char == '/':
                stack.append(int(operand_1 / operand_2))
    
    return stack.pop()

postfix_expression = "82+4*"
result = evaluate_postfix(postfix_expression)
print("Postfix expression:", postfix_expression)
print("Result:", result)
```

```python
# 10.  Implement a queue using the stack data structure

class queue:
    def __init__(self):
        self.data = []

    # adding the data
    def enqueue(self, element):
        self.data.append(element)

    # removing first element
    def dequeue(self):
        if not self.isEmpty():
            return self.data.pop(0)
        else:
            return None

    # return first element
    def front(self):
        if not self.isEmpty():
            return self.data[0]
        else:
            return None
    
    # return last element
    def rear(self):
        if not self.isEmpty():
            return self.data[-1]
        else:
            return None

    def display(self):
        print(self.data)

    def size(self):
        return len(self.data)

    def isEmpty(self):
        return self.size() == 0

queue = queue()
queue.enqueue(1)
queue.enqueue(3)
queue.enqueue(5)
queue.display()

print(queue.front())
print(queue.rear())
print(queue.isEmpty())
print(queue.size())
```
